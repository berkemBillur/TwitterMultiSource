```python
## Part of stage 2 to grab the necessary info of the multi-source case.
# grab the base information of the infectors of the tweet.

import json 
import os
from types import new_class
import pandas as pd
from copy import deepcopy
import tweepy
import time
from datetime import datetime
import os.path
import random
from math import floor

class crawler(object):
    def __init__(self, hashtag): 

        self.hashtag = hashtag
        self.sep = os.path.sep
        self.path = f'tweets/{hashtag}'
    ################################################################################################################
    ################################################################################################################
    ################################################################################################################


####################################################################################

# FUNCTINOS INVOLVING GETTING THE TWEETS WITH HASHTAGS

####################################################################################

    def load_new_informer_df(self):
        path = f'{self.path}/updated_informers_data.json'
        with open(path) as jf:
            data = json.load(jf)
        return data

    def load_new_ms(self):
        path = f'{self.path}/{self.hashtag}_ms_cases.json'
        with open(path) as jf:
            data = json.load(jf)
        return data


    # function to find the multi-source cases
    def find_ms(self):
        print(f'\n{self.hashtag}')

        data = self.load_new_informer_df()
        df = pd.DataFrame.from_dict(data,orient='index')

        ms = df[ (df['infector-is-friend']==1) &(df['num-informers']>1)]
        ms_d = ms.to_dict(orient='index')

        for key,value in ms_d.items():
            if value['num-informers'] > 10:
                value['informers-data'] = random.sample(value['informers-data'],10)
                # value['informers-data'] = value['informers-data'][0::int(value['num-informers']/10)]
                value['num-informers'] = len(value['informers-data'])

        save = f'{self.path}/{self.hashtag}_ms_cases.json'
        with open(save, 'w') as fp:
            json.dump(ms_d, fp)
```

```python
hashtags = ['avengers','blm','borisjohnson','brexit','climatechange','covid','gaza','loveisland','monkeypox','nhs','olivianewtonjohn','supercup','UkraineWar']

for hashtag in hashtags:
      
    c = crawler(hashtag)
    data = c.find_ms()
```
