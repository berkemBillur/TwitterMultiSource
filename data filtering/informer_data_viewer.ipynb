```python
import json 
import os
import pandas as pd

hashtags = ['Casemiro']
# ,'gaza','avengers','brexit','borisjohnson','blm']

for hashtag in hashtags:


    path = f'{hashtag}{os.path.sep}updated_informers_data.json'
    with open(path) as jf:
        data = json.load(jf)

    for key, value in data.items():
        num_inf = len(value['informers-data'])
        value['num-informers'] = num_inf

    df = pd.DataFrame.from_dict(data,orient='index')

    # df.to_csv(f'{hashtag}{os.path.sep}inf_db_pres.csv')
```
