```python
hashtags = ['avengers','blm','brexit','climatechange','covid','gaza','loveisland','monkeypox','nhs','olivianewtonjohn','supercup','UkraineWar']


cats = ['sentiment', 'topic', 'topic_single', 'irony', 'emoji', 'offensive','gender']

sentiment_df = []

polite_grammar_df = []

polarity_subjec_df = []

emotion_df = []

offensive_df = []

# lets get the average stats for each hashtag
for i, hashtag in enumerate(hashtags):

    df = pd.read_csv(f'multisource_analysis/user_ft_data/{hashtag}_scored_tweets2.csv')
    df.drop(['Unnamed: 0', 'target','is-target','is-infector','is-informer',
             'text','hashtag'], inplace = True, axis =1)

    cols = df.columns.to_list()
    [ cols.remove(x) for x in cols if 'count' in x ]
    cols = [ x for x in cols if 'user' in x]

    sent_cols = ['user_neg_mean','user_pos_mean','user_neu_mean']

    polite_grammar_cols = ['user_politeness_mean','user_grammar_score_mean']

    polarity_subjec_cols = ['user_polarity_mean','user_subjectivity_mean']

    emotion_cols = ['user_joy_mean', 'user_sadness_mean', 'user_others_mean', 
                     'user_anger_mean', 'user_surprise_mean', 'user_disgust_mean', 
                    'user_fear_mean']

    offensive_cols = ['user_targeted_mean','user_hateful_mean','user_aggressive_mean']


    fig1 = df.copy()[sent_cols].mean() )

    fig1 = df.copy()[polite_grammar_cols].mean() )

    fig1 = df.copy()[polarity_subjec_cols].mean() )

    fig1 = df.copy()[emotion_cols].mean() )

    fig1 = df.copy()[offensive_cols].mean() )



    print(f'added {hashtag} to consideration')
```
